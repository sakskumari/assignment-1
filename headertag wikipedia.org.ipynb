```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
# header tag from wikipedia.org

header_tags = soup.find(class_="mw-headline")
header_tags.text
```
